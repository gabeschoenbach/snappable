```python
complete_df = pd.read_csv("heuristics/complete_10000.csv", index_col=0).rename(columns={"proportion_snaps": "Complete Graph"})[1:]
square_df = pd.read_csv("heuristics/square_10000.csv", index_col=0).rename(columns={"proportion_snaps": "Square Grid Graph"})

complete_edges_index = [0.5 * ((n ** 2) - n) for n in complete_df.index]
square_edges_index = [2 * n * (n - 1) for n in square_df.index]
square_vertices_index = [index ** 2 for index in square_df.index]

square_df = square_df.set_index(pd.Index(square_vertices_index))
```

```python
fig, ax = plt.subplots(figsize=(16,4))
logx = True

complete_df.plot(ax=ax,
                 marker='o',
                 markersize=10,
                 logx=logx,
                 )
square_df.plot(ax=ax,
               marker='o',
               markersize=10,
               logx=logx,
              )

plt.xlabel("|V|", fontsize=24)
# plt.ylabel("Proportion", fontsize=18)
# _ = plt.title("Proportion of Snappable Trees", fontsize=24)
ax.tick_params(axis='x', labelsize=18)
ax.tick_params(axis='y', labelsize=18)
os.makedirs("heuristics/plots", exist_ok=True)
plt.savefig("heuristics/plots/proportion_snaps_V.png", dpi=300, bbox_inches='tight')
```
